## Definição das *RNA´s*:

* Uma rede neural é um modelo matemático inspirado no funcionamento do cérebro humano.  

* É uma estrutura composta por unidades de processamento chamadas de **neurônios artificiais**, que estão interconectados em camadas. 

* Cada neurônio recebe entradas, realiza cálculos e produz uma saída que é transmitida para outros neurônios na rede.

10-2150972x10.jpg

## Conceitos Principais:

* **Neurônio Artificial**: O neurônio artificial é a unidade básica de processamento em uma rede neural. Ele recebe um conjunto de entradas ponderadas, aplica uma função de ativação e produz uma saída. O neurônio artificial é inspirado no neurônio biológico do cérebro humano.

* **Camadas**: Uma rede neural é composta por camadas de neurônios. A camada de entrada recebe os dados de entrada, enquanto as camadas ocultas processam as informações intermediárias, e a camada de saída gera as previsões ou classificações finais. Cada neurônio em uma camada está conectado a todos os neurônios na camada seguinte.

* **Ponderação e Bias**: Cada conexão entre neurônios tem um peso associado, que determina a importância daquela conexão na computação do neurônio de destino. Além disso, cada neurônio tem um bias, que é um valor constante adicionado à soma ponderada das entradas. Os pesos e biases são ajustados durante o treinamento da rede neural.

* **Função de Ativação**: A função de ativação é aplicada à soma ponderada das entradas de um neurônio e determina a saída desse neurônio. Ela introduz a não-linearidade na rede neural, permitindo que ela aprenda relações complexas nos dados. Exemplos de funções de ativação comuns incluem a função sigmoid, a função ReLU e a função softmax.

## Lendo o Banco de Dados:

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

## Pré-Processamento:

```python
x_train = x_train.reshape((60000, 784)) / 255.0
x_test = x_test.reshape((10000, 784)) / 255.0
```

## Processamento:

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Criar o modelo da rede neural
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))
```

```python
# Compilar o modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Treinar o modelo
model.fit(x_train, y_train, epochs=5, batch_size=128, validation_data=(x_test, y_test))
```

## Resultado:

```python
# Avaliar o modelo
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Acurácia do modelo: {accuracy}")
```
