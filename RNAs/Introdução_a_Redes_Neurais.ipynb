## Definição das *RNA´s*:

* Uma rede neural artificial é um modelo matemático inspirado no funcionamento do cérebro humano.  

* Cada neurônio recebe entradas, realiza cálculos e produz uma saída que é transmitida para outros neurônios na rede.

image.png

> ***Perceptron*** é uma rede neural de camada única e um Perceptron de várias camadas é chamado de Rede Neural Artificial. O Perceptron é um classificador linear (binário). Além disso, é usado na aprendizagem supervisionada e pode ser usado para classificar os dados de entrada fornecidos.

image.png

**Potencial de Ativação:**

```
y = f(w1*x1 + w2*x2 + ... + wn*xn + b)
```

Nesta fórmula:

* y é a saída do perceptron (0 ou 1, dependendo da função de ativação utilizada).
* x1, x2, ..., xn são os valores das entradas.
* w1, w2, ..., wn são os pesos sinápticos correspondentes às entradas.
* b é o bias (também conhecido como viés) do perceptron.
* f() é a função de ativação.

**Função de ativação de f():**

```
f(z) = {
    1, se z >= 0,
    0, caso contrário
}
```

## Conceitos Principais:

* **Neurônio Artificial**: O neurônio artificial é a unidade básica de processamento em uma rede neural. Ele recebe um conjunto de entradas ponderadas, aplica uma função de ativação e produz uma saída. O neurônio artificial é inspirado no neurônio biológico do cérebro humano.

* **Camadas**: Uma rede neural é composta por camadas de neurônios. A camada de entrada recebe os dados de entrada, enquanto as camadas ocultas processam as informações intermediárias, e a camada de saída gera as previsões ou classificações finais. Cada neurônio em uma camada está conectado a todos os neurônios na camada seguinte.

* **Ponderação e Bias**: Cada conexão entre neurônios tem um peso associado, que determina a importância daquela conexão na computação do neurônio de destino. Além disso, cada neurônio tem um bias, que é um valor constante adicionado à soma ponderada das entradas. Os pesos e biases são ajustados durante o treinamento da rede neural.

* **Pesos Sinápticos**: Os pesos sinápticos são valores atribuídos a cada conexão entre neurônios em uma rede neural. Esses pesos determinam a importância relativa de cada entrada para o neurônio de destino. Cada conexão sináptica é multiplicada pelo seu respectivo peso antes de ser somada no neurônio de destino. Os pesos sinápticos são ajustados durante o treinamento da rede para melhorar o desempenho da rede em uma determinada tarefa.

* **Combinador Linear**: O combinador linear é uma operação matemática que calcula a soma ponderada das entradas de um neurônio. Essa soma ponderada é obtida multiplicando cada entrada pelo seu respectivo peso sináptico e somando os resultados. O combinador linear é usado para agregar as informações das entradas antes de serem processadas pela função de ativação do neurônio. O resultado do combinador linear é passado para a função de ativação, que introduz a não-linearidade na rede neural.

* **Limiar de Ativação**: O limiar de ativação é um valor limite que determina se um neurônio será ativado ou não. É uma propriedade individual de cada neurônio. Após a soma ponderada das entradas no combinador linear, o resultado é comparado ao limiar de ativação. Se o valor do combinador linear for maior ou igual ao limiar, o neurônio será ativado e produzirá uma saída; caso contrário, ele permanecerá inativo.

* **Potencial de Ativação**: O potencial de ativação é o resultado da soma ponderada das entradas no combinador linear de um neurônio antes de ser processado pela função de ativação. É o valor obtido multiplicando cada entrada pelo seu respectivo peso sináptico e somando os resultados.

* **Função de Ativação**: A função de ativação é aplicada ao potencial de ativação de um neurônio para determinar sua saída final. Ela introduz a não-linearidade na rede neural e permite que ela aprenda e represente relações complexas nos dados. Existem várias funções de ativação comumente usadas em redes neurais, incluindo**:

  * **Função Sigmoid**: Também conhecida como função logística, a função sigmoid mapeia o potencial de ativação para um valor entre 0 e 1. É definida pela fórmula f(x) = 1 / (1 + exp(-x)). É frequentemente usada em problemas de classificação binária.

  * **Função ReLU (Rectified Linear Unit)**: A função ReLU retorna o valor do potencial de ativação se for maior que zero e zero caso contrário. Matematicamente, é definida por f(x) = max(0, x). A função ReLU é amplamente usada em redes neurais devido à sua simplicidade computacional e à capacidade de lidar com o problema do desaparecimento do gradiente.

  * **Função Softmax**: A função softmax é usada na camada de saída de uma rede neural quando se lida com problemas de classificação multiclasse. Ela mapeia o potencial de ativação para uma distribuição de probabilidade sobre as diferentes classes, garantindo que a soma das saídas seja igual a 1.

## Lendo o Banco de Dados:

````python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
````

## Pré-Processamento:

````python
x_train = x_train.reshape((60000, 784)) / 255.0
x_test = x_test.reshape((10000, 784)) / 255.0
````

## Processamento:

````python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
````

````python
# Criar o modelo da rede neural
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))
````

````python
# Compilar o modelo
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
````

````python
# Treinar o modelo
model.fit(x_train, y_train, epochs=5, batch_size=128, validation_data=(x_test, y_test))
````

* ***Epoch:*** Uma época representa uma passagem completa dos dados de treinamento pela rede neural. Durante uma época, os pesos da rede são atualizados com base no algoritmo de otimização usado (como o Gradiente Descendente). O número de épocas especificado durante o treinamento determina quantas vezes os dados de treinamento serão percorridos pela rede.

* ***469/469:*** Nesse exemplo, indica que há um total de 469 lotes (batches) de dados de treinamento. Cada lote contém um conjunto de amostras de treinamento processadas em paralelo. Esses lotes são usados para calcular as atualizações dos pesos durante o treinamento.

* ***loss:*** É a função de perda (loss function), que mede a diferença entre as previsões da rede neural e os rótulos (labels) verdadeiros. O objetivo durante o treinamento é minimizar essa função de perda, ajustando os pesos da rede.

* ***accuracy:*** É a métrica de precisão (accuracy), que representa a porcentagem de amostras de treinamento que foram corretamente classificadas pela rede neural. É uma medida de desempenho que indica o quão bem a rede está se saindo durante o treinamento. Nesse exemplo, a precisão é reportada para os dados de treinamento.

* ***val_loss:*** É a função de perda calculada nos dados de validação. Durante o treinamento, é comum dividir os dados em um conjunto de treinamento e um conjunto de validação separados. A função de perda nos dados de validação é uma medida do desempenho da rede em dados não vistos durante o treinamento. O objetivo é que essa função de perda também seja minimizada.

* ***val_accuracy:*** É a métrica de precisão calculada nos dados de validação. Ela representa a porcentagem de amostras de validação que foram corretamente classificadas pela rede neural.

## Resultado:

````python
# Avaliar o modelo
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Acurácia do modelo: {accuracy}")
````

## Pratica

#### configuração inical

````python
# Definindo local de armazenamento dos modelos 
import os
MODELS_DIR = 'models/'
if not os.path.exists(MODELS_DIR):
    os.mkdir(MODELS_DIR)
MODEL_TF = MODELS_DIR + 'model'
MODEL_NO_QUANT_TFLITE = MODELS_DIR + 'model_no_quant.tflite'
MODEL_TFLITE = MODELS_DIR + 'model.tflite'
MODEL_TFLITE_MICRO = MODELS_DIR + 'model.cc'
````

````python
# TensorFlow é uma biblioteca de aprendizado de máquina de código aberto
import tensorflow as tf

# Keras é a API de alto nível do TensorFlow para aprendizado profundo
from tensorflow import keras
# Numpy é uma biblioteca matemática
import numpy as np
# Pandas é uma biblioteca de manipulação de dados
import pandas as pd
# Matplotlib é uma biblioteca de gráficos
import matplotlib.pyplot as plt
# Math é a biblioteca matemática do Python
import math

# Define a semente para reprodutibilidade do experimento
seed = 1
np.random.seed(seed)
tf.random.set_seed(seed)
````

#### gerando exemplo de dados

````python
# Número de pontos de dados de amostra
SAMPLES = 1000

# Gere um conjunto de números aleatórios distribuídos uniformemente no intervalo de
# 0 a 2π, que abrange uma oscilação completa de uma onda senoidal
x_values = np.random.uniform(low=0, high=2*math.pi, size=SAMPLES).astype(np.float32)

# Embaralhe os valores para garantir que não estejam em ordem
np.random.shuffle(x_values)

# Calcule os valores de seno correspondentes
y_values = np.sin(x_values).astype(np.float32)

# Adicione um pequeno número aleatório a cada valor de y
y_values += 0.1 * np.random.randn(*y_values.shape)

# Usaremos 60% dos nossos dados para treinamento, 20% para teste e os 20% restantes
# para validação. Calcule os índices de cada seção.
TRAIN_SPLIT = int(0.6 * SAMPLES)
TEST_SPLIT = int(0.2 * SAMPLES + TRAIN_SPLIT)

# Use np.split para dividir nossos dados em três partes.
# O segundo argumento para np.split é uma matriz de índices onde os dados serão
# divididos. Fornecemos dois índices, então os dados serão divididos em três partes.
x_train, x_test, x_validate = np.split(x_values, [TRAIN_SPLIT, TEST_SPLIT])
y_train, y_test, y_validate = np.split(y_values, [TRAIN_SPLIT, TEST_SPLIT])
````

#### Treinando modelo e gerando modelo

````python
model = tf.keras.Sequential()

# A primeira camada recebe uma entrada escalar e a alimenta através de 16 "neurônios". Os
# neurônios decidem se ativam com base na função de ativação 'relu'.
model.add(keras.layers.Dense(16, activation='relu', input_shape=(1,)))

# A segunda e terceira camadas ajudarão a rede a aprender representações mais complexas
model.add(keras.layers.Dense(16, activation='relu'))

# A camada final é um único neurônio, pois queremos produzir um único valor de saída
model.add(keras.layers.Dense(1))

# Compile o modelo usando o otimizador padrão 'adam' e a função de perda de erro quadrático médio 'mse' para regressão.
model.compile(optimizer='adam', loss="mse", metrics=["mae"])
````

````python
# Treinando modelo
history = model.fit(x_train, y_train, epochs=500, batch_size=64, validation_data=(x_validate, y_validate))

# Salvando model
model.save(MODEL_TF)
````

#### Fazendo conversão para modelo tiny

````python
# Converter o modelo para o formato TensorFlow Lite sem quantização
converter = tf.lite.TFLiteConverter.from_saved_model(MODEL_TF)
model_no_quant_tflite = converter.convert()

# Salvar o modelo no disco
open(MODEL_NO_QUANT_TFLITE, "wb").write(model_no_quant_tflite)

# Converter o modelo para o formato TensorFlow Lite com quantização
def representative_dataset():
    for i in range(500):
        yield [x_train[i].reshape(1, 1)]
        
# Definir a flag de otimização.
converter.optimizations = [tf.lite.Optimize.DEFAULT]
# Forçar quantização somente para inteiros
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
# Fornecer um conjunto de dados representativo para garantir a quantização correta.
converter.representative_dataset = representative_dataset
model_tflite = converter.convert()

# Salvar o modelo no disco
open(MODEL_TFLITE, "wb").write(model_tflite)
````
