```python
import requests
from bs4 import BeautifulSoup as soup

url = 'https://loterias.caixa.gov.br/Paginas/Mega-Sena.aspx'
html = requests.get(url).text
soupPage = soup(html, 'html5lib') 

#ul = soupPage.findAll( "ul", id="ulDezenas" ) #retorna array com a tag UL com a classe informada
li = soupPage.find("ul", class_="numbers megasena").findAll("li")
```
